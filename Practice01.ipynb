```python
# Add total sales column
df["Total"] = df["Quantity"] * df["Price"]

# Total revenue
print("Total Revenue:", df["Total"].sum())

# Top-selling product
print(df.groupby("Product")["Quantity"].sum().sort_values(ascending=False))

# Customer who spent the most
print(df.groupby("Customer")["Total"].sum().sort_values(ascending=False))

# Revenue by product
print(df.groupby("Product")["Total"].sum())

# Daily revenue
print(df.groupby("Date")["Total"].sum())
```

**Practice to clean text. let say we want to read data from web**

```python
states = ["   Alabama ", "Georgia!", "Georgia", "georgia", "FlOrIda",
          "south   carolina##", "West virginia?"]
```

```python
import re # importing regular expression to do a complex canipulation

def clean_strings(strings): #input values
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub("[!#?]", "", value)
        value = value.title()
        result.append(value)
    return result

cleaned = clean_strings(states)

print (cleaned)
```
