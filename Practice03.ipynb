```python
import nltk # import the package
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize

text = "This is an example sentence, showing off the tokenization process."

tokens = word_tokenize(text)

print(tokens)

# ['This', 'is', 'an', 'example', 'sentence', ',', 'showing', 'off', 'the', 'tokenization', 'process', '.']
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng') # pos tagging

from nltk.tokenize import word_tokenize
from nltk import pos_tag

text = "This is an example sentence, showing off the part-of-speech tagging process."

tokens = word_tokenize(text)
tagged_tokens = pos_tag(tokens)

print(tagged_tokens)
```

Lets create a full grammer

```python
import nltk
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('maxent_ne_chunker_tab')


from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk

text = "Barack Obama was born in Hawaii, and he was the President of the United States."

tokens = word_tokenize(text)
tagged_tokens = pos_tag(tokens)
ne_chunked_tokens = ne_chunk(tagged_tokens)

print(ne_chunked_tokens)
```
